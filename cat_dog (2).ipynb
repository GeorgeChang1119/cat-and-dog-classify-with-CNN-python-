```python
# training data前處理
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import pandas as pd
training_data=[]
DATADIR='C:/pythonpratice/training_set/training_set/'
CATEGORIES=['dogs','cats']
def create_training_data():                    
    for category in CATEGORIES:
        path=os.path.join(DATADIR,category)  #path to cats or dogs dir    # os.path.join
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):                # os.listdir
            img_array=cv2.imread(os.path.join(path, img))       
            new_array=cv2.resize(img_array,(100,100))
            training_data.append([new_array,class_num])
            

create_training_data()
```

```python
np.array(training_data).shape
```

```python
import random

random.shuffle(training_data)       #將 list內容順序打亂
```

```python
x_train=[]
y_train=[]
```

```python
for features, label in training_data:
    x_train.append(features)
    y_train.append(label)
x_train=np.array(x_train)
y_train=np.array(y_train)
```

```python
x_train.shape,y_train.shape
```

```python
x_train=x_train/255.0
x_train[0]
```

```python
# test data前處理
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import pandas as pd
test_data=[]
DATADIR='C:/pythonpratice/test_set/test_set/'
CATEGORIES=['dogs','cats']
def create_test_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR,category)  #path to cats or dogs dir
        class_num=CATEGORIES.index(category)
        for img in os.listdir(path):
            img_array=cv2.imread(os.path.join(path, img))
            new_array=cv2.resize(img_array,(100,100))
            test_data.append([new_array,class_num])

create_test_data()
```

```python
len(test_data)
```

```python
import random

random.shuffle(test_data)   
```

```python
test_data[9]
```

```python
x_test=[]
y_test=[]
```

```python
for features, label in test_data:
    x_test.append(features)
    y_test.append(label)
x_test=np.array(x_test)
y_test=np.array(y_test)
```

```python
x_test.shape,y_test.shape
```

```python
x_test=x_test/255.0 
x_test[0]
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, AveragePooling2D  #2D只會向右向下掃描
 
keras.backend.clear_session()
np.random.seed(1)
tf.random.set_seed(1)
```

```python
model=Sequential([
    Conv2D(filters=30, kernel_size=(10,10),strides=(2,2),padding='valid',activation='relu',input_shape=(100,100,3),kernel_initializer='he_normal'),
    Dropout(rate=0.25),
    MaxPooling2D(pool_size=(5,5),padding='valid'),          #stides預設同 poolsize         #kernel_size為filter之大小
    
    Conv2D(filters=60,kernel_size=5,strides=4,padding='same',activation='relu',kernel_initializer='he_normal'),    #size=2同size=(2,2)
    Dropout(rate=0.2),
    AveragePooling2D(pool_size=3, padding='same'),
    
    Flatten(),
    Dropout(rate=0.25),
    Dense(units=1024,activation='relu',kernel_initializer='he_normal'),
    Dropout(rate=0.25),
    Dense(units=2, activation='softmax')   
])
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy']) #SGD之優化法
```

```python
train=model.fit(x_train, y_train, validation_split=0.2, epochs=25, batch_size=128, verbose=1)
```

```python
pd.DataFrame(train.history).plot(figsize=(8,8))
plt.grid()
plt.show()
```

```python
model.evaluate(x_test, y_test, verbose=1)
#model.evaluate(x_test, y_test)
```

```python
y_proba= model.predict(x_test)
y_proba[:3].round(2)
```

```python
y_pred=np.argmax(y_proba,axis=1)
y_pred[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
